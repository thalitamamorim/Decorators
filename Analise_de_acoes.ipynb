```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel('/content/Analise de dados.xlsx', sheet_name='Principal')
df_principal
```

```python
df_total_acoes = pd.read_excel('/content/Analise de dados.xlsx', sheet_name='Total_de_acoes')
df_total_acoes
```

```python
df_ticker = pd.read_excel('/content/Analise de dados.xlsx', sheet_name='Ticker')
df_ticker
```

```python
df_chat = pd.read_excel('/content/Analise de dados.xlsx', sheet_name='ChatGTP')
df_chat
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final','Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['var_pct'] = df_principal['var_dia_pct']/100
df_principal['valor_inicial'] = df_principal['valor_final']/(df_principal['var_pct']+1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format='{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtde_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['variacao_rs'].apply(lambda x: 'Subiu' if x>0 else ('Desceu' if x<0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chat, left_on='Nome', right_on='Empresa', how='left')
df_principal = df_principal.drop(columns=['Empresa'])
df_principal
```

```python
pd.options.display.float_format = '{:,.2f}'.format

media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['variacao_rs'].mean(skipna=True)
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['variacao_rs'].mean(skipna=True)

print(f"Maior: \tR$ {maior:,.2f}")
print(f"Menor: \tR$ {menor:,.2f}")
print(f"Média: \tR$ {media:,.2f}")
print(f"Média de quem Subiu: \tR$ {media_subiu:,.2f}")
print(f"Média de quem Desceu: \tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
def formatar_reais(numero):
    return '{:,.2f}'.format(numero).replace(',', '|').replace('.', ',').replace('|', '.')

maior_formatado = formatar_reais(maior)
menor_formatado = formatar_reais(menor)
media_formatada = formatar_reais(media)
media_subiu_formatada = formatar_reais(media_subiu)
media_desceu_formatada = formatar_reais(media_desceu)

print(f"Maior: \tR$ {maior_formatado}")
print(f"Menor: \tR$ {menor_formatado}")
print(f"Média: \tR$ {media_formatada}")
print(f"Média de quem Subiu: \tR$ {media_subiu_formatada}")
print(f"Média de quem Desceu: \tR$ {media_desceu_formatada}")
```

```python
df_analise_saldo['variacao_rs'] = df_analise_saldo['variacao_rs'].round(2)

def formatar_reais(valor):
    return f'R$ {valor:,.2f}'.replace('.', '|').replace(',', '.').replace('|', ',')

fig = px.bar(df_analise_saldo,
             x='Resultado',
             y='variacao_rs',
             text=df_analise_saldo['variacao_rs'].apply(formatar_reais),
             title='Variação Reais por Resultado')
fig.show()
```

```python
def formatar_reais(valor):
    return 'R${:,.2f}'.format(valor).replace(',', '|').replace('.', ',').replace('|', '.')

df_analise_cat_idade = df_principal.groupby('Idade (anos)')['variacao_rs'].sum().reset_index()
df_analise_cat_idade['variacao_rs'] = df_analise_cat_idade['variacao_rs'].apply(formatar_reais)

fig = px.bar(df_analise_cat_idade,
             x='Idade (anos)',
             y='variacao_rs',
             text='variacao_rs',
             title='Variação Reais por Categoria de Idade')

fig.show()
```

```python
fig = px.pie(df_analise_segmento, names='Segmento', values='variacao_rs', title='Variação Reais por Segmento')
fig.show()
```
